```python
import csv
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

def Plot(filename,xAxis,yAxis):
    with open(filename) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        total_count = 0
        headers = next(csv_reader)
        print(headers)
        noOfRows = 5000     #Set -1 to display all columns        
        xaxis = [row[headers.index(xAxis)] for row in csv_reader][0:noOfRows]        
        csv_file.seek(0)
        headers = next(csv_reader)
        yaxis = [float(row[headers.index(yAxis)]) for row in csv_reader][0:noOfRows]        
        fig, ax = plt.subplots()
        plt.gcf().autofmt_xdate()
        ax.plot(xaxis, yaxis)
        ax.bar(xaxis, yaxis)
        plt.show()
        
        
if __name__=="__main__":
   # Plot("covid19_cases.csv",'dateRep','cases')
    Plot("covid19_cases.csv",'countriesAndTerritories','cases')
    
```
